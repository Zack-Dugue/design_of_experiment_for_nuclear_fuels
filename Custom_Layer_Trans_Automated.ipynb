```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras import layers as KL
```

```python
RADIUS_IDX = 12

class PhysicsInverseSquareGate(KL.Layer):
    """
    Physics gate that scales the static memory using w(r) = alpha / (r + eps)^p.
    - radius_idx: which column in X_static is the distance r (0-based)
    - learn_alpha/learn_power: if False, fixes alpha=1, p=2 (pure inverse-square)
    - residual_blend: learn a gamma in [0,1] to softly mix raw vs. gated features
    """
    def __init__(self, radius_idx, learn_alpha=True, learn_power=True,
                 eps=5e-4, clip_max=1.0, residual_blend=True, **kwargs):
        super().__init__(**kwargs)
        self.radius_idx = int(radius_idx)
        self.learn_alpha = learn_alpha
        self.learn_power = learn_power
        self.eps = float(eps)
        self.clip_max = clip_max
        self.residual_blend = residual_blend

    def build(self, input_shapes):
        # softplus^-1(1.0) ≈ 0.5413249 so softplus(alpha_raw) starts at ~1.0
        if self.learn_alpha:
            self.alpha_raw = self.add_weight(
                name="alpha_raw", shape=(), initializer=tf.keras.initializers.Constant(0.5413249), trainable=True
            )
        else:
            self.alpha_raw = None

        # p = 1 + softplus(p_raw); with p_raw ≈ 0.5413 -> p ≈ 2.0 init
        if self.learn_power:
            self.p_raw = self.add_weight(
                name="p_raw", shape=(), initializer=tf.keras.initializers.Constant(0.5413249), trainable=True
            )
        else:
            self.p_raw = None

        if self.residual_blend:
            # sigmoid(1.3863) ≈ 0.8 → start with 80% gated, 20% identity
            self.gamma_raw = self.add_weight(
                name="gamma_raw", shape=(), initializer=tf.keras.initializers.Constant(1.3862944), trainable=True
            )
        super().build(input_shapes)

    def call(self, inputs):
        x_static, z_static = inputs                     # x_static: [B,S], z_static: [B,S,d]
        r = x_static[:, self.radius_idx]                # [B]
        r = tf.math.abs(r)                              # ensure non-negative

        alpha = tf.nn.softplus(self.alpha_raw) if self.learn_alpha else tf.constant(1.0, z_static.dtype)
        p     = 1.0 + tf.nn.softplus(self.p_raw) if self.learn_power else tf.constant(2.0, z_static.dtype)

        w = alpha / tf.pow(r + self.eps, p)             # [B]
        if self.clip_max is not None:
            w = tf.clip_by_value(w, 0.0, self.clip_max)
        w = tf.reshape(w, (-1, 1, 1))                   # [B,1,1] (broadcast over tokens and d_model)

        gated = z_static * w                            # [B,S,d]

        if self.residual_blend:
            gamma = tf.sigmoid(self.gamma_raw)          # scalar in (0,1)
            return (1.0 - gamma) * z_static + gamma * gated
        return gated
```

```python
# ===================================================
# 🔹 Data Augmentation: Add Gaussian noise to training inputs
# ===================================================
def augment_with_noise(X, alpha=0.001):
    """
    Adds Gaussian noise scaled by each feature's standard deviation.
    alpha = strength of noise (0.01 = small, 0.05 = moderate)
    """
    std = np.std(X, axis=0)
    noise = np.random.normal(0, alpha * std, X.shape)
    return X + noise
```

```python
# =========================
# End-to-end Transformer (static X + windowed y), ONE decoder layer
# =========================

# --- hyperparams
T_TOTAL   = 72
SEQ_LEN   = 8          # window length L
STRIDE    = 2           # quarter overlap
S_TOKENS  = 14          # number of static features (tokens)
D_MODEL   = 128
N_HEADS   = 8
D_FF      = 4 * D_MODEL
DROPOUT   = 0.01
BATCH     = 32
LR        = 0.001
eps      =  LR
clip_max    = 5.0
assert D_MODEL % N_HEADS == 0


# -------------------------------------------------------------------
# Windowing + tf.data
# -------------------------------------------------------------------
def window_starts(total_len=T_TOTAL, L=SEQ_LEN, stride=STRIDE):
    starts = list(range(0, total_len - L + 1, stride))
    last = total_len - L
    if starts[-1] != last:
        starts.append(last)
    return starts

STARTS = window_starts()  # [0,8,16,24,32,40]

def build_y_windows(y_2d, L=SEQ_LEN, starts=STARTS):
    """
    y_2d: (N, T_TOTAL) -> y_in, y_tgt, tids
      y_in  : (N*num_w, L, 1)  shifted-right with BOS=0 at index 0
      y_tgt : (N*num_w, L, 1)
      tids  : (N*num_w, L)     absolute time-step IDs in [0..71]
    """
    N, T = y_2d.shape
    num_w = len(starts)
    y_in  = np.zeros((N * num_w, L, 1), dtype=np.float32)
    y_tgt = np.zeros((N * num_w, L, 1), dtype=np.float32)
    tids  = np.zeros((N * num_w, L),   dtype=np.int32)
    k = 0
    for n in range(N):
        yy = y_2d[n]
        for t0 in starts:
            y_tgt[k, :, 0] = yy[t0:t0+L]
            y_in[k, 1:, 0] = yy[t0:t0+L-1]
            y_in[k, 0, 0]  = 0.0
            tids[k, :]     = np.arange(t0, t0+L)
            k += 1
    return y_in, y_tgt, tids

def repeat_x_for_windows(X, num_windows):
    return np.repeat(X.astype(np.float32), num_windows, axis=0)

def make_ds_e2e(X_static, y_in, tids, y_tgt, batch=BATCH, shuffle=True):
    X_static = X_static.astype(np.float32)
    y_in     = y_in.astype(np.float32)
    y_tgt    = y_tgt.astype(np.float32)
    tids     = tids.astype(np.int32)      # ONLY tids must be ints

    ds = tf.data.Dataset.from_tensor_slices(((X_static, y_in, tids), y_tgt))
    if shuffle:
        ds = ds.shuffle(len(y_in), seed=42, reshuffle_each_iteration=True)
    return ds.batch(batch).prefetch(tf.data.AUTOTUNE)


#=========================================================================================
#                  Noise Function 
# ===================================================
# 🔹 Data Augmentation (only for training)
# ===================================================
X_train_aug = augment_with_noise(X_train, alpha=0.001)
print("✅ Added Gaussian noise to training data:", X_train_aug.shape)

# Repeat augmented static features for windowed samples
X_tr_rep = repeat_x_for_windows(X_train_aug, len(STARTS))

#=========================================================================================


# Build datasets (assumes X_train/test & y_train/test already exist)
y_in_tr, y_tgt_tr, tids_tr = build_y_windows(y_train)
y_in_te, y_tgt_te, tids_te = build_y_windows(y_test)
#X_tr_rep = repeat_x_for_windows(X_train, len(STARTS))
X_te_rep = repeat_x_for_windows(X_test,  len(STARTS))

train_ds = make_ds_e2e(X_tr_rep, y_in_tr, tids_tr, y_tgt_tr, shuffle=True)
test_ds  = make_ds_e2e(X_te_rep, y_in_te, tids_te, y_tgt_te, shuffle=False)

# quick dtype sanity
(Xb, Yin_b, tids_b), Yt_b = next(iter(train_ds))
print("dtypes:", Xb.dtype, Yin_b.dtype, Yt_b.dtype, tids_b.dtype)  # float32 float32 float32 int32

# -------------------------------------------------------------------
# Blocks
# -------------------------------------------------------------------
class AddFeatureIDEmbedding(KL.Layer):
    """Learned feature-ID embedding: adds a vector E[i] to token i."""
    def __init__(self, num_features, d_model, **kwargs):
        super().__init__(**kwargs)
        self.E = self.add_weight(
            (num_features, d_model),
            initializer="glorot_uniform", trainable=True
        )
    def call(self, tokens):  # tokens: [B,S,d_model]
        return tokens + self.E

def build_static_encoder_block(num_feats=S_TOKENS, d_model=D_MODEL,
                               n_heads=N_HEADS, d_ff=D_FF, dropout=DROPOUT):
    x_in = KL.Input(shape=(num_feats,), dtype=tf.float32, name="X_static")
    x = KL.Lambda(lambda t: tf.expand_dims(t, -1), name="expand_last")(x_in)    # [B,S,1]
    x = KL.Dense(d_model, name="scalar_to_dmodel")(x)                           # [B,S,d_model]
    x = AddFeatureIDEmbedding(num_feats, d_model, name="add_feat_id")(x)        # [B,S,d_model]
    # self-attention
    attn = KL.MultiHeadAttention(n_heads, d_model//n_heads, dropout=dropout,
                                 name="enc_self_attn")(x, x)
    y = KL.Dropout(dropout, name="enc_attn_do")(attn)
    y = KL.Add(name="enc_attn_resid")([x, y])
    y = KL.LayerNormalization(axis=-1, epsilon=1e-6, name="enc_attn_ln")(y)
    # FFN
    f = KL.Dense(d_ff, activation=tf.nn.gelu, name="enc_ffn1")(y)
    f = KL.Dropout(dropout, name="enc_ffn_do1")(f)
    f = KL.Dense(d_model, name="enc_ffn2")(f)
    f = KL.Dropout(dropout, name="enc_ffn_do2")(f)
    z = KL.Add(name="enc_ffn_resid")([y, f])
    z = KL.LayerNormalization(axis=-1, epsilon=1e-6, name="enc_ffn_ln")(z)
    return Model(x_in, z, name="StaticEncoder")



#============================================================================================================================================================
#=============================================================================================================================================================
##  Sinusoidal Porisional Encoding
# ==== 1) Sinusoidal positional encoding layer (fixed, non-trainable) ====
class SinusoidalPositionalEncoding(KL.Layer):
    """
    Vaswani et al. 2017 sinusoidal PE.
    Inputs:
      tids: int32 tensor [B, L] with absolute (or relative) time indices.
    Outputs:
      pe: float32 tensor [B, L, d_model]
    """
    def __init__(self, d_model, max_len, **kwargs):
        super().__init__(**kwargs)
        assert d_model % 2 == 0, "D_MODEL must be even for sinusoidal PE."
        self.d_model = d_model
        self.max_len = max_len

    def build(self, input_shape):
        # [max_len]
        pos = tf.range(self.max_len, dtype=tf.float32)[:, None]          # [max_len, 1]
        i = tf.range(self.d_model // 2, dtype=tf.float32)[None, :]       # [1, d_model/2]
        div_term = tf.pow(10000.0, (2.0 * i) / tf.cast(self.d_model, tf.float32))  # [1, d_model/2]

        angles = pos / div_term                                          # [max_len, d_model/2]
        pe = tf.concat([tf.sin(angles), tf.cos(angles)], axis=1)         # [max_len, d_model]
        self.pe = tf.Variable(pe, trainable=False, name="sinusoid_table")
        super().build(input_shape)

    def call(self, tids):  # tids: [B, L] int32
        # gather rows for each timestep id
        return tf.gather(self.pe, tids)  # [B, L, d_model]





##=========================================================================================================================================================
##=========================================================================================================================================================
#♥5♣#♥5===================================================================================================================================================

def build_decoder_masked_self_attn(seq_len=SEQ_LEN, d_model=D_MODEL,
                                   n_heads=N_HEADS, dropout=DROPOUT, t_total=T_TOTAL):
    y_in  = KL.Input(shape=(seq_len, 1), dtype=tf.float32, name="y_in")
    tids  = KL.Input(shape=(seq_len,),   dtype=tf.int32,   name="time_ids")
    ytok  = KL.Dense(d_model, name="y_scalar_to_dmodel")(y_in)
    temb = SinusoidalPositionalEncoding(d_model=d_model, max_len=t_total, name="time_sinusoid")(tids)             #temb = KL.Embedding(input_dim=t_total, output_dim=d_model, name="time_abs_emb")(tids)
    ytok  = KL.Add(name="add_time_emb")([ytok, temb])
    mha   = KL.MultiHeadAttention(n_heads, d_model//n_heads, dropout=dropout,
                                  name="dec_masked_self_attn")(ytok, ytok, use_causal_mask=True)
    h     = KL.Dropout(dropout, name="dec_attn_do")(mha)
    h     = KL.Add(name="dec_attn_resid")([ytok, h])
    h     = KL.LayerNormalization(axis=-1, epsilon=1e-6, name="dec_attn_ln")(h)
    return Model([y_in, tids], h, name="DecoderMSA")

def build_decoder_redbox(seq_len=SEQ_LEN, s_tokens=S_TOKENS, d_model=D_MODEL,
                         n_heads=N_HEADS, d_ff=D_FF, dropout=DROPOUT):
    dec_states_in = KL.Input(shape=(seq_len, d_model), dtype=tf.float32, name="dec_states_in")
    static_mem_in = KL.Input(shape=(s_tokens, d_model), dtype=tf.float32, name="static_mem_in")
    # cross-attention
    cross = KL.MultiHeadAttention(n_heads, d_model//n_heads, dropout=dropout,
                                  name="dec_cross_attn")(dec_states_in, static_mem_in)
    x = KL.Dropout(dropout, name="dec_cross_dropout")(cross)
    x = KL.Add(name="dec_cross_residual")([dec_states_in, x])
    x = KL.LayerNormalization(axis=-1, epsilon=1e-6, name="dec_cross_add_norm")(x)
    # FFN
    f = KL.Dense(d_ff, activation=tf.nn.gelu, name="dec_ffn_dense_1")(x)
    f = KL.Dropout(dropout, name="dec_ffn_dropout_1")(f)
    f = KL.Dense(d_model, name="dec_ffn_dense_2")(f)
    f = KL.Dropout(dropout, name="dec_ffn_dropout_2")(f)
    y = KL.Add(name="dec_ffn_residual")([x, f])
    y = KL.LayerNormalization(axis=-1, epsilon=1e-6, name="dec_ffn_add_norm")(y)
    # linear head (regression)
    y_hat = KL.Dense(1, name="out_head")(y)
    return Model([dec_states_in, static_mem_in], y_hat, name="DecoderRedBox")

# -------------------------------------------------------------------
# Assemble end-to-end model
# Function to automate Transformers-based model.

def build_transformer_e2e():
    static_encoder = build_static_encoder_block()
    decoder_msa   = build_decoder_masked_self_attn()
    redbox        = build_decoder_redbox()

    X_in   = KL.Input(shape=(S_TOKENS,),   dtype=tf.float32, name="X_static_in")
    y_in_i = KL.Input(shape=(SEQ_LEN, 1),  dtype=tf.float32, name="y_in_in")
    tids_i = KL.Input(shape=(SEQ_LEN,),    dtype=tf.int32,   name="tids_in")

    Z_static_raw = static_encoder(X_in)      # (B, 14, D_MODEL)

    # ---- Physics gate on static memory ----
    Z_static = PhysicsInverseSquareGate(
        radius_idx=RADIUS_IDX,
        learn_alpha=True,
        learn_power=True,
        eps=eps,
        clip_max=clip_max,
        residual_blend=True,
        name="physics_gate"
    )([X_in, Z_static_raw])

    dec_states = decoder_msa([y_in_i, tids_i])   # (B, L, D_MODEL)
    y_hat      = redbox([dec_states, Z_static])  # (B, L, 1)

    model = Model([X_in, y_in_i, tids_i], y_hat, name="Transformer_E2E")
    model.compile(
        optimizer=tf.keras.optimizers.Adam(LR),
        loss="mse",
        metrics=["mae"]
    )
    return model

```

```python
# Use the same constants you trained with
T_TOTAL = 72
SEQ_LEN = SEQ_LEN    # L you trained with

@tf.function(experimental_relax_shapes=True)
def _predict_window(model, X1, y_in1, tids1):
    # X1: (1, 14)  y_in1: (1, L, 1)  tids1: (1, L)
    return model([X1, y_in1, tids1], training=False)  # (1, L, 1)

def autoregressive_predict_full72(model, X_unseen, seq_len=SEQ_LEN, t_total=T_TOTAL):
    """
    Args:
        model: trained end-to-end Keras model (X, y_in, tids) -> (L,1)
        X_unseen: (N_unseen, 14) static features (float32, normalized)
    Returns:
        Y_pred: (N_unseen, 72) predictions
    """
    X_unseen = X_unseen.astype(np.float32)
    N = X_unseen.shape[0]
    Y_pred = np.zeros((N, t_total), dtype=np.float32)

    for i in range(N):
        x = X_unseen[i:i+1]  # (1, 14)
        y_gen = np.zeros((t_total,), dtype=np.float32)

        for t in range(t_total):
            # choose start index so the window is valid and contains the last (seq_len-1) values
            t0 = max(0, min(t - (seq_len - 1), t_total - seq_len))
            rel = t - t0  # index inside the window we care about

            # build inputs for this window
            y_in_win = np.zeros((1, seq_len, 1), dtype=np.float32)
            if rel > 0:
                # teacher-forcing with our own past predictions (shifted right by 1 with BOS at [0])
                y_in_win[0, 1:rel+1, 0] = y_gen[t0:t]

            tids_win = np.arange(t0, t0 + seq_len, dtype=np.int32)[None, :]  # (1, L)

            # run the trained model to get predictions for the whole window
            y_win_hat = _predict_window(model, x, y_in_win, tids_win).numpy()  # (1, L, 1)

            # take the value for the current time step t (relative position `rel`)
            y_gen[t] = y_win_hat[0, rel, 0]

        Y_pred[i] = y_gen

    return Y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

N_RUNS   = 1
BASE_SEED = 1234
results  = []

for run in range(N_RUNS):
    seed = BASE_SEED + run
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

    print(f"\n===== RUN {run+1}/{N_RUNS} — seed {seed} =====")

    # -------- build fresh model for this run --------
    model = build_transformer_e2e()

    # ------------------- train ----------------------
    history = model.fit(
        train_ds,
        validation_data=test_ds,
        epochs=300,
        callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_mae", patience=40, restore_best_weights=True)],
        verbose=1
    )
    #=======================================================================================
    # --------------- test (interpolation) -----------
    # Predict on full validation set
    y_pred = model.predict([X_te_rep, y_in_te, tids_te], batch_size=32)

    test_mae = val_mae_full = mean_absolute_error(y_tgt_te.flatten(), y_pred.flatten())
    test_loss = val_mae_full = mean_squared_error(y_tgt_te.flatten(), y_pred.flatten())
    mae = test_mae          # MAE metric
    mse = test_loss         # MSE loss
    #=============================================================================
    # ----------------- unseen metrics ---------------
    # --- Run inference on unseen set ---
    Y_pred_unseen = autoregressive_predict_full72(model, X_unseen)

    # --- Evaluate (normalized scale) ---
    mae_unseen = np.mean(np.abs(Y_pred_unseen - y_unseen))
    mse_unseen = np.mean((Y_pred_unseen - y_unseen) ** 2)
    # ----------------- save this run ----------------
    results.append({
        "run": run + 1,
        "seed": seed,
        "mae": mae,
        "mse": mse,
        "unseen_mae": mae_unseen,
        "unseen_mse": mse_unseen
    })

# ---------------- save to CSV ----------------
df = pd.DataFrame(results)
df.to_csv("UO2_0.711_VFX_PhysTrans.csv", index=False)
print(df)

print("\nMAE mean/std:", df["mae"].mean(), df["mae"].std())
print("MSE mean/std:", df["mse"].mean(), df["mse"].std())
print("Unseen MAE mean/std:", df["unseen_mae"].mean(), df["unseen_mae"].std())
print("Unseen MSE mean/std:", df["unseen_mse"].mean(), df["unseen_mse"].std())
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# USER SETTINGS (EDIT THESE)
# ============================================================
FUEL_CSV_FOLDER = "C:/Users/49757884/Desktop/Gustavo/Doctoral_Research/DOE/New_Transformer/HGR/fuel"  # <-- CHANGE
TIME_COL_START  = 8
T_TOTAL         = 72
SEQ_LEN         = 8
STRIDE          = 2
ROW_IDX         = 33  # sample (row) you want to plot

# ============================================================
# 1) ELAPSED DAYS DICT FROM CSV FOLDER
# ============================================================
def build_fuel_elapsed_days_dict(folder_path, time_col_start=8, max_days=72):
    file_paths = [
        os.path.join(folder_path, f)
        for f in os.listdir(folder_path)
        if f.lower().endswith(".csv")
    ]
    d = {}
    for fp in file_paths:
        df = pd.read_csv(fp)
        time_cols = df.columns[time_col_start:]
        try:
            elapsed_days = np.array([float(c) for c in time_cols], dtype=float)
        except ValueError:
            print(f"[WARN] Non-numeric elapsed-day header in: {os.path.basename(fp)}")
            continue

        elapsed_days = elapsed_days[:max_days]
        fuel_name = os.path.splitext(os.path.basename(fp))[0]
        d[fuel_name] = elapsed_days

    if not d:
        raise RuntimeError("fuel_elapsed_days_dict is empty. Check folder path and TIME_COL_START.")
    print(f"Loaded elapsed-days keys: {list(d.keys())}")
    return d

# ============================================================
# 2) FUEL TYPE APPROXIMATION
# ============================================================
KNOWN_FUELS = {
    "UC_0.711_RB":   [0.711, 0.806548, 3],
    "UN_0.711_RB":   [0.711, 1.0,      3],
    "UO2_0.35_RB":   [0.35,  0.0,      3],
    "UO2_0.35_VXF":  [0.35,  0.0,      1],
    "UO2_0.711_RB":  [0.711, 0.0,      3],
    "UO2_0.711_VXF": [0.711, 0.0,      1],
}

def approximate_fuel_type(row_3vals, known_fuels=KNOWN_FUELS):
    target = np.asarray(row_3vals, dtype=float)
    best = min(known_fuels, key=lambda k: np.linalg.norm(np.asarray(known_fuels[k], dtype=float) - target))
    return best

# Optional: normalize for filename mismatches (UO2 vs U02 etc.)
def normalize_fuel_name(s: str) -> str:
    s = s.upper()
    s = s.replace("UO2", "U02")   # <-- if your CSV filenames use U02
    s = s.replace(" ", "")
    s = s.replace("-", "_")
    return s

def find_matched_key(fuel_elapsed_days_dict, fuel_type_row):
    fuel_type_norm = normalize_fuel_name(fuel_type_row)
    matched_key = next(
        (k for k in fuel_elapsed_days_dict.keys()
         if fuel_type_norm in normalize_fuel_name(k)),
        None
    )
    return matched_key

# ============================================================
# 3) RAS POSITION DECODER (same indexing assumption as before)
# ============================================================
def get_ras_position_from_row(row):
    row = np.asarray(row)

    col3_encoded_val = row[2]
    col4_encoded_val = row[3]
    col5_val = int(row[4])
    col6_onehot = row[5:]  # one-hot starts here (adjust if needed)

    col3_reverse_map = {1: "VXF", 3: "RB"}
    col3_str = col3_reverse_map.get(int(col3_encoded_val), "UNK")

    col4_reverse_lookup = {
        ("RB",  3): (1, 2),
        ("RB",  2): (5, 3),
        ("RB",  1): (4, 1),
        ("VXF", 2): (2, 3),
        ("VXF", 1): (1, 1),
    }
    col4_tuple = col4_reverse_lookup.get((col3_str, int(col4_encoded_val)), ("?", "?"))
    col6_digit = int(np.argmax(col6_onehot))

    ras_str = f"{col3_str}[{','.join(map(str, col4_tuple))}]{col5_val}"
    return ras_str, col6_digit

# ============================================================
# 4) BLOCK STITCHING + ROW FEATURE PICK (handles repeated X)
# ============================================================
def stitch_overlap_blocks(blocks, seq_len=SEQ_LEN, stride=STRIDE):
    b = np.asarray(blocks)
    if b.ndim == 3 and b.shape[-1] == 1:
        b = b[..., 0]
    if b.ndim != 2 or b.shape[1] != seq_len:
        raise ValueError(f"Expected (n_blocks,{seq_len}) or (n_blocks,{seq_len},1). Got {b.shape}")

    y = list(b[0])
    for blk in b[1:]:
        y.extend(blk[-stride:])
    return np.asarray(y)

def blocks_per_row(t_total=T_TOTAL, seq_len=SEQ_LEN, stride=STRIDE):
    return (t_total - seq_len) // stride + 1

def get_row_feature_vector(X_features, row_idx, bpr):
    """
    Works for:
      - per-sample X: (n_samples, n_features)
      - block-repeated X: (n_samples*bpr, n_features)
    """
    X = np.asarray(X_features)
    if X.shape[0] >= (row_idx + 1) * bpr:
        return X[row_idx * bpr]  # first block's features for that row
    return X[row_idx]

# ============================================================
# 5) THE "OTHER" PLOTTING FUNCTION (toggles)
# ============================================================
def plot_row_blocks_elapsed_days(
    y_pred_blocks, y_true_blocks,
    X_features,
    row_idx,
    fuel_elapsed_days_dict,
    t_total=T_TOTAL, seq_len=SEQ_LEN, stride=STRIDE,
    show_point_numbers=False,      # <-- toggle point labels
    show_metrics_in_title=False,   # <-- toggle MAE/MSE in title
    annotate_every=1,
    save_pdf_path=None
):
    bpr = blocks_per_row(t_total, seq_len, stride)
    start = row_idx * bpr
    end   = start + bpr

    y_pred = stitch_overlap_blocks(y_pred_blocks[start:end], seq_len=seq_len, stride=stride)
    y_true = stitch_overlap_blocks(y_true_blocks[start:end], seq_len=seq_len, stride=stride)

    if len(y_pred) != t_total or len(y_true) != t_total:
        raise ValueError(f"Stitched length mismatch: pred={len(y_pred)} true={len(y_true)} expected={t_total}")

    row_feat = get_row_feature_vector(X_features, row_idx, bpr)

    # fuel type (uses first 3 values)
    fuel_type = approximate_fuel_type([row_feat[0], row_feat[1], row_feat[2]])

    matched_key = find_matched_key(fuel_elapsed_days_dict, fuel_type)
    if matched_key is None:
        raise ValueError(
            f"No mapping found: fuel_type_row='{fuel_type}'. "
            f"Check KNOWN_FUELS keys vs CSV filenames."
        )

    elapsed_days = np.asarray(fuel_elapsed_days_dict[matched_key]).ravel()[:t_total]
    if len(elapsed_days) != t_total:
        raise ValueError(f"Elapsed-days mismatch: got {len(elapsed_days)} expected {t_total} for key={matched_key}")

    ras_position, col6_value = get_ras_position_from_row(row_feat)

    mae = float(np.mean(np.abs(y_pred - y_true)))
    mse = float(np.mean((y_pred - y_true) ** 2))

    # ---- title construction (toggle metrics) ----
    title = f"RAS: {ras_position}"
    if show_metrics_in_title:
        title += f"\nMAE={mae:.3f}  MSE={mse:.3f}"

    # ---- plot ----
    plt.figure(figsize=(10, 6))
    plt.plot(elapsed_days, y_true, marker="o", label="Actual")
    plt.plot(elapsed_days, y_pred, marker="x", label="Prediction")
    plt.xlabel("Irradiation Days", fontsize=16)
    plt.ylabel("Prediction HGR (W/g)", fontsize=16)
    plt.title(title,  fontsize=18)
    plt.tick_params(axis="both", which="major", labelsize=14)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # ---- point labels (toggle) ----
    if show_point_numbers:
        for i, (xi, yi) in enumerate(zip(elapsed_days, y_true)):
            if i % annotate_every != 0:
                continue
            plt.annotate(
                f"{xi:g}",
                (xi, yi),
                textcoords="offset points",
                xytext=(0, 6),
                ha="center",
                fontsize=7,
                alpha=0.85
            )

    if save_pdf_path:
        plt.savefig(save_pdf_path, format="pdf", bbox_inches="tight")

    plt.show()

# ============================================================
# 6) RUN
# ============================================================
fuel_elapsed_days_dict = build_fuel_elapsed_days_dict(
    FUEL_CSV_FOLDER,
    time_col_start=TIME_COL_START,
    max_days=T_TOTAL
)


# Example call (UNCOMMENT and replace variables):
plot_row_blocks_elapsed_days(
    y_pred_blocks=y_pred_blocks,
    y_true_blocks=y_tgt_te,
    X_features=X_te_rep,
    row_idx=ROW_IDX,
    fuel_elapsed_days_dict=fuel_elapsed_days_dict,
    show_point_numbers=False,       # <-- no numbers on points
    show_metrics_in_title=False,    # <-- no MAE/MSE in title
    annotate_every=1,
    save_pdf_path= "C:\\Users\\49757884\\Desktop\\Gustavo\\Doctoral_Research\\DOE\\AA_Result\\Inter_HGR_UO2(0.711VXF).pdf"
)
```

```python
# Extract known second-column values from your fuel_signatures
known_fuels = {
    "UC_0.711_RB":[0.711, 0.806548, 3], 
    "UN_0.711_RB": [0.711, 1.0, 3],
    "UO2_0.35_RB": [0.35, 0.0, 3],
    "UO2_0.35_VXF":[0.35, 0.0, 1],
    "UO2_0.711_RB":[0.711, 0.0, 3],
    "UO2_0.711_VXF":[0.711, 0.0, 1]
}
def approximate_fuel_type(row, known_fuels):
    # row = [col0_scaled, col1_scaled, col3_encoded]
    target_vector = np.array(row)

    closest_fuel = min(
        known_fuels,
        key=lambda k: np.linalg.norm(np.array(known_fuels[k]) - target_vector)
    )
    return closest_fuel

approx_fuel_types = [
    approximate_fuel_type(row, known_fuels) for row in X_full_pred_df[['col0', 'col1_scaled', 'col3_encoded']].values
]
```

```python
def get_ras_position_from_unseen(X_full_pred_df, row_idx, encoder):
    row = X_full_pred_df[row_idx]

    # Indices based on your X construction:
    col3_encoded_val = row[2]  # col3_encoded
    col4_encoded_val = row[3]  # col4_encoded
    col5_val = int(row[4])     # original col5
    col6_onehot = row[5:]      # One-hot part starts here
    
    # --- Reverse col3_encoded ---
    col3_reverse_map = {1: 'VXF', 3: 'RB'}
    col3_str = col3_reverse_map.get(int(col3_encoded_val), 'UNK')

    # --- Reverse col4_encoded ---
    # Approximate reverse mapping for illustration
    col4_reverse_lookup = {
        ('RB', 3): (1, 2),
        ('RB', 2): (5, 3),
        ('RB', 1): (4,),
        ('VXF', 2): (2, 3),
        ('VXF', 1): (1,)
    }
    col4_tuple = col4_reverse_lookup.get((col3_str, int(col4_encoded_val)), ('?',))

    # --- Recover col6 digit from one-hot ---
    col6_digit = int(np.argmax(col6_onehot))  # inverse of one-hot encoding

    # --- Construct final string ---
    ras_str = f"{col3_str}({','.join(map(str, col4_tuple))}){col5_val}"

    return ras_str, col6_digit
```
